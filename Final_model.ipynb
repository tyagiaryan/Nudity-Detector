```python
def nudity_model(url):
    model = load_model("nudity_final.h5")
    urllib.request.urlretrieve(url, "output.jpg")
    img=image.load_img('output.jpg',target_size=(224,224))
    plt.imshow(img)
    plt.show()
    x=image.img_to_array(img)
    x=np.expand_dims(x,axis=0)
    images=np.vstack([x])
    pred=model.predict(images,batch_size=1) 
    if pred[0][0]>0.5:
        category = "NonNude"
    elif pred[0][1]>0.5:
        category = "Nude"    
    print(category)
```
